# Loading the dataframe

## Data Preprocessing

```python
#Checking for null values
null_counts = df.isna().sum()
print(null_counts)
```

```python
#Checking for duplicated rows
total_duplicates = df.duplicated().sum()
print(f'Total duplicated rows: {total_duplicates}')
```

```python
#Droping the duplicated rows
df.drop_duplicates(inplace = True)
df.shape
```

```python
#Change column name Diabetes_binary to Diabetes (the Outcome)
df['Diabetes'] = df['Diabetes_binary']
df.drop(columns = 'Diabetes_binary', inplace = True)
```

```python
#Value count for each feature
for x in df.columns:
    print(x,'\n',df[x].value_counts())
    print('-'*30)
```

```python
#Change dtype to int32
df = df.astype('int32')
```

## Exploratory Data Analysis
### General analysis

```python
#Summary statistics
summary_stats = df.describe()
print(summary_stats)
```

```python
#Visualize summary statistics using a boxplot
plt.figure(figsize = (10, 6))
sns.boxplot(data = df, orient = 'h')
plt.title('Summary Statistics - Boxplot')
plt.show()
```

```python
#Correlation between the variables and the outcome (diabetes)
df.corr()['Diabetes'][:-1].sort_values().plot(kind='bar')
```

```python
#Separate dataframes for diabetics and non-diabetics
df_diabetic = df[df['Diabetes'] == 1]
df_non_diabetic = df[df['Diabetes'] == 0]
```

### BMI

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_diabetic['BMI'], color='blue', label='Diabetic', kde=True)
sns.histplot(df_non_diabetic['BMI'], color='red', label='Non-Diabetic', kde=True)
plt.title("Histogram of BMI by Diabetes Status")
plt.xlabel("BMI")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,5))
sns.boxplot(x='Diabetes', y='BMI', data=df, ax=ax[0]).set_title('BMI vs Diabetes')
sns.violinplot(x='Diabetes', y='BMI', data=df, ax=ax[1]).set_title('BMI vs Diabetes')
```

Non diabetic patients have a normal BMI within the range of 25-35 whereas the diabetic patients have a greater BMI.

### Age

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_diabetic['Age'], color='blue', label='Diabetic', kde=False)
sns.histplot(df_non_diabetic['Age'], color='red', label='Non-Diabetic', kde=False)
plt.title("Histogram of Age by Diabetes Status")
plt.xlabel("Age Group | 1 = 18-24 ... 9 = 60-64 ... 13 = 80 or older")
plt.ylabel("Density")
plt.legend()
plt.show()
```

This histogram shows the distribution of diabetics and non-diabetics by age. Elderly people are at high risk of diabetes.

### Sex

```python
# Calculate the count of male and female diabetes patients
male_patients = df_diabetic[df_diabetic['Sex'] == 1]['Sex'].count()
female_patients = df_diabetic[df_diabetic['Sex'] == 0]['Sex'].count()

# Create a pie chart
labels = ['Male', 'Female']
sizes = [male_patients, female_patients]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # Explode the 1st slice (Male)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Distribution of Diabetes Patients by Sex")
plt.axis('equal')

plt.show()
```

There seems to be almost no relation between diabetes and sex since the distribution is almost the same.

### High cholesterol

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Bar Chart for the Relationship between HighChol and Diabetes
sns.countplot(data=df, x='HighChol', hue='Diabetes', ax=axes[0])
axes[0].set_title("Relationship between High Cholesterol and Diabetes")
axes[0].set_xlabel("High Cholesterol | 0 = no  1 = yes")
axes[0].set_ylabel("Count")
axes[0].legend(title="Diabetes", labels=["No Diabetes", "Diabetes"])

# Bar Chart for High Cholesterol in Diabetics
sns.countplot(data=df[df['Diabetes'] == 1], x='HighChol', ax=axes[1])
axes[1].set_title("Distribution of High Cholesterol in Diabetics")
axes[1].set_xlabel("High Cholesterol | 0 = no  1 = yes")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
df_diabetic['HighChol'].value_counts().sort_values(ascending=False)
```

Both plots show the relation between the High Cholesterol and Diabetes. Among diabetics 23496 have high cholesterol, while 11601 have cholesterol in a normal range.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Bar Chart for the Relationship between Difficulty in Walking and Diabetes
sns.countplot(data=df, x='DiffWalk', hue='Diabetes', ax=axes[0])
axes[0].set_title("Relationship between Difficulty in walking and Diabetes")
axes[0].set_xlabel("Difficulty in Walking | 0 = no  1 = yes")
axes[0].set_ylabel("Count")
axes[0].legend(title="Diabetes", labels=["No Diabetes", "Diabetes"])

# Bar Chart for Difficulty in Walking in Diabetics
sns.countplot(data=df[df['Diabetes'] == 1], x='DiffWalk', ax=axes[1])
axes[1].set_title("Distribution of the Difficulty in Walking in Diabetics")
axes[1].set_xlabel("Difficulty in Walking | 0 = no  1 = yes")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

Although some diabetics do not present difficulty in walking or climbing stairs, there is a noticeable proportion of diabetics that present this feature.
